```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import SelectKBest, chi2, mutual_info_classif
```

```python
df=pd.read_csv("Telco_Customer_Churn_Dataset  (1).csv")
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.columns)
```

```python
df.drop('customerID', axis=1, inplace=True, errors='ignore')
```

```python
if 'TotalCharges' in df.columns:
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df.dropna(inplace=True)
```

```python
le = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])
```

```python
X = df.drop('Churn', axis=1)
y = df['Churn']
```

```python
chi_selector = SelectKBest(score_func=chi2, k=10)
chi_selector.fit(X, y)
```

```python
chi_scores = pd.DataFrame({
    'Feature': X.columns,
    'Chi-Square Score': chi_selector.scores_
}).sort_values(by='Chi-Square Score', ascending=False)

print("Top Features (Chi-Square):\n")
print(chi_scores.head(10))
```

```python
mi_scores = mutual_info_classif(X, y)
mi_df = pd.DataFrame({
    'Feature': X.columns,
    'MI Score': mi_scores
}).sort_values(by='MI Score', ascending=False)

print("\nTop Features (Mutual Information):\n")
print(mi_df.head(10))
```
